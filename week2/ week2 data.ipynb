```python
def func1(row):
    if '기저귀' in row:
        return '기저귀'
    if '팬티' in row:
        return '팬티'
    if '젖병' in row:
        return '수유'
    if '하기스' in row:
        return '기저귀'
    if '젖꼭지' in row:
        return '수유'
#   return row
    else:
       return '기타'

df1['상품중분류'] = df1["상품명"].apply(func1)
df1['상품중분류'].value_counts()
```

```python
df1.head()
```

```python
df1['배송시작일(dt)'] = pd.to_datetime(df1['배송시작일'])
df1['배송완료일(dt)'] = pd.to_datetime(df1['배송완료일'])
```

```python
df1['배송기간'] = df1['배송완료일(dt)'] - df1['배송시작일(dt)']
```

```python
df1.describe()
```

```python
df1_pivot = df1.pivot_table( index='물품대분류',
                             values='결제금액',aggfunc='sum').reset_index()
df1_pivot_sort = df1_pivot.sort_values(by='결제금액',ascending=False)
df1_pivot_sort
```

```python
df1_pivot_sort['매출비중'] = df1_pivot_sort['결제금액']/df1_pivot_sort['결제금액'].sum()
df1_pivot_sort
```

```python
df1['물품대분류'].value_counts()
```

```python
sns.barplot(data=df1_pivot_sort, x='물품대분류',y='결제금액')
```

```python
df1['구매일(dt)'] = pd.to_datetime(df1['구매일'])
```

```python
df1_pivot2 = df1.pivot_table(index=['구매일(dt)','물품대분류'],
                             values='결제금액',aggfunc='sum').reset_index()
df1_pivot2
```

```python
plt.figure(figsize=[20,5])
sns.lineplot(data=df1_pivot2, x='구매일(dt)',y='결제금액', ci=None)
```

```python
plt.figure(figsize=[20,15])
sns.lineplot(data=df1_pivot2, x='구매일(dt)',y='결제금액', ci=None, hue='물품대분류')
```

```python
cond1 = (df1_pivot2['구매일(dt)'] >= '2020-01-01')
df1_pivot3 = df1_pivot2.loc[cond1]
```
